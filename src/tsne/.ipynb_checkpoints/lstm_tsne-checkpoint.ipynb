# 使用 lstm 抽取出來的 feature 來進行視覺化

## import package

```python
import torch
import ast
import csv
import random
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

## set parameters

```python
BATCH_SIZE = 128
LATENT_DIM = 512
REDUCED_DIM = 300
NUM_ITER = 10000
MODEL_NAME = "../lstm/model.pth"
DATA_PATH = "../lstm/data/dt_all.csv"
pca = False
perplexity = 150
lr = 5e-4

DEVICE_ID = 0
SEED = 5566

torch.cuda.set_device(DEVICE_ID)
torch.cuda.set_per_process_memory_fraction(0.8)
use_gpu = torch.cuda.is_available()
device = torch.device("cuda" if use_gpu else "cpu")

torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)
random.seed(SEED)
np.random.seed(SEED)
```

## funtion to load training data

```python
def load_train_label(df, train_dt, label_dt):
    train_dt = ["dt%d" % i for i in train_dt]
    label_dt = ["dt%d" % (label_dt - 1), "dt%d" % (label_dt)]
    label = df[label_dt].values.tolist()
    idx = df['chid'].tolist()
    tag_feat = df[train_dt].values.tolist()
    for i in range(len(tag_feat)):
        tag_feat[i] = np.array([ast.literal_eval(x) for x in tag_feat[i]])
    for j in range(len(label)):
        label[j][1] = np.array(ast.literal_eval(label[j][1]))

    return idx, tag_feat, [i[1] for i in label]
```

```python
class LSTM_Dataset(torch.utils.data.Dataset):
    def __init__(self, id_list, tag, labels):
        self.id_list = id_list
        self.tag = tag
        self.labels = labels
    
    def __getitem__(self, idx):
        return self.id_list[idx], self.tag[idx], self.labels[idx]
    
    def __len__(self):
        return len(self.id_list)
```

## define model

```python
class LSTM_Backbone(torch.nn.Module):
    def __init__(self, hidden_dim, num_layers, bidirectional, fix_embedding=True):
        super(LSTM_Backbone, self).__init__()
        self.lstm = torch.nn.LSTM(16, hidden_dim, num_layers=num_layers, \
                                  bidirectional=bidirectional, batch_first=True, dropout = 0.55)
        
    def forward(self, inputs):
        x, _ = self.lstm(inputs)

        return x
    
class Header(torch.nn.Module):
    def __init__(self, dropout, hidden_dim):
        super(Header, self).__init__()
        self.classifier = torch.nn.Sequential(torch.nn.Dropout(dropout),
                                  torch.nn.Linear(hidden_dim, 16),
                                  torch.nn.Sigmoid())
    
    def forward(self, inputs):
        inputs = inputs.sum(dim=1)
        out = self.classifier(inputs)

        return out
```

## function for dimension reduction (TSNE)

```python
def dim_reduction(backbone, device, test_loader, n_iter, reduced_dim, pca, perplexity):
    backbone.eval()
    backbone.to(device)
    with torch.no_grad():
        latent_vecs = torch.tensor([]).to(device, dtype = torch.float)
        labels = torch.tensor([])
        for idx, (idx, tags, label) in enumerate(test_loader):
            tags = tags.float().to(device)
            if not backbone is None:
                latent_vec = backbone(tags)
                latent_vec = latent_vec.sum(dim = 1)
            latent_vecs = torch.cat((latent_vecs, latent_vec), dim = 0)
            labels = torch.cat((labels, label))
            if (len(latent_vecs) > 5000):
                break

        latent_vecs = latent_vecs.cpu().detach().numpy()

    if (pca):
        pca = PCA(n_components = reduced_dim, copy = False, whiten = True, svd_solver = 'full')
        latent_vecs = pca.fit_transform(latent_vecs)

    tsne = TSNE(n_components = 2, verbose = 1, method = 'exact', perplexity = perplexity, n_iter = n_iter)
    latent_vecs = tsne.fit_transform(latent_vecs)

    return latent_vecs, labels
```

## draw the visualization result for each class in a 4 x 4 grid

```python
def draw_each(data_embedded, targets):
    subplot_kw = {"xticks": [], "yticks": []}
    gridspec_kw = {"wspace": 0.1, "hspace": 0.2}
    fig, axs = plt.subplots(4, 4, gridspec_kw = gridspec_kw, subplot_kw = subplot_kw,
                            figsize = (12, 9))

    for tag in range(16):
        i = int(tag / 4)
        j = tag % 4
        axs[i, j].set_title(f"tag {tag}", {"fontsize": 6})
        tag_embedded = data_embedded[targets == tag]
        axs[i, j].scatter(tag_embedded[:, 0], tag_embedded[:, 1], s = 0.5, c = COLORS[tag])

    plt.show()

    return
```

## load data and create dataloader object

```python
data = pd.read_csv(DATA_PATH)
test_idx, test_tag, test_label = load_train_label(data, [i for i in range(15, 23)], 23)
test_dataset = LSTM_Dataset(test_idx, test_tag, test_label)
test_loader = torch.utils.data.DataLoader(dataset = test_dataset,
                                            batch_size = 128,
                                            shuffle = False)
```

## apply TSNE

```python
checkpoint = torch.load(MODEL_NAME)
backbone = LSTM_Backbone(hidden_dim=256, num_layers=4, bidirectional=True)
backbone.load_state_dict(checkpoint["backbone"])
 
latent_vecs, labels = dim_reduction(backbone, device, test_loader, NUM_ITER,
                                   reduced_dim = REDUCED_DIM, pca = pca, perplexity = perplexity)
```

## parameters for drawing the visualization result

```python
targets = np.argmax(labels.numpy(), axis = 1)
COLORS = ["steelblue", "red", "black", "brown", "yellow", "peachpuff", "cyan", "darkgray",
          "greenyellow", "hotpink", "lime", "magenta", "olivedrab", "orange", "blue", "peru"]
vecs_colors = np.array([COLORS[target] for target in targets])
```

## draw all 16 classes in 2d

```python
plt.scatter(latent_vecs[:, 0], latent_vecs[:, 1], s = 3, c = vecs_colors)
plt.show()
```

## draw each class seperately

```python
for target in range(16):
    latent_vecs_target = latent_vecs[tuple([targets == target])]
    plt.scatter(latent_vecs_target[:, 0], latent_vecs_target[:, 1], s = 3, c = COLORS[target])
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

## draw the visualization result for each class in a 4 x 4 grid

```python
draw_each(latent_vecs, targets)
```
